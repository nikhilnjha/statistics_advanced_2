# Theory

1. What is hypothesis testing in statistics
  - Hypothesis testing is a method to make statistical decisions using sample data.
  - It assesses whether evidence exists to reject a specific assumption about a population (null hypothesis).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis
  - The null hypothesis (H₀) claims no effect or difference.
  - The alternative hypothesis (H₁) suggests there is an effect or difference.

3. What is the significance level in hypothesis testing, and why is it important
  - The significance level (α, often 0.05) is the probability threshold for rejecting H₀.
  - It defines the risk of making a Type 1 error—wrongly rejecting a true null hypothesis.

4. What does a P-value represent in hypothesis testing
  - The P-value indicates the probability of observing your result, or more extreme, under the null hypothesis.

5. How do you interpret the P-value in hypothesis testing
  - If P ≤ α: reject H₀ (significant).
  - If P > α: fail to reject H₀ (not significant).

6. What are Type 1 and Type 2 errors in hypothesis testing
  - Type 1 Error: Rejecting a true H₀ (false positive).
  - Type 2 Error: Failing to reject a false H₀ (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing
  - One-tailed tests check for a difference in one direction.
  - Two-tailed tests check for any difference in both directions.

8. What is the Z-test, and when is it used in hypothesis testing
  - The Z-test is used when population variance is known and sample size is large (n > 30).
  - It tests if a sample mean significantly differs from a known population mean.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing
        Z = (X̄ - μ) / (σ / √n)
  - It shows how many standard deviations a sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution
  - The T-distribution is used when the sample size is small and population std. dev. is unknown.
  - It adjusts for extra variability from smaller samples.

11. What is the difference between a Z-test and a T-test
  - Z-test: population std. dev. known, large sample.
  - T-test: std. dev. unknown, small sample.

12. What is the T-test, and how is it used in hypothesis testing
  - A T-test compares means to assess if they differ significantly, especially with small sample sizes.

13. What is the relationship between Z-test and T-test in hypothesis testing
  - Both test means. T-test is preferred for small n. As n increases, the T-distribution approximates the Z.

14. What is a confidence interval, and how is it used to interpret statistical results
  - A confidence interval gives a range where the population parameter likely lies.
  - It communicates estimation uncertainty.

15. What is the margin of error, and how does it affect the confidence interval
  - The margin of error quantifies estimation error.
  - Larger margin → wider interval → less precise estimate.

16. How is Bayes' Theorem used in statistics, and what is its significance
  - Bayes' Theorem updates probabilities with new evidence.
  - It's key in Bayesian inference and decision-making.

17. What is the Chi-square distribution, and when is it used
  - The Chi-square distribution is used for categorical data to assess differences in observed vs. expected counts.

18. What is the Chi-square goodness of fit test, and how is it applied
  - This test checks if sample categorical data matches an expected distribution using frequency comparisons.

19. What is the F-distribution, and when is it used in hypothesis testing
  - The F-distribution compares variances. Used in ANOVA and regression model evaluation.

20. What is an ANOVA test, and what are its assumptions
  - ANOVA compares means across 3+ groups. Assumes:
      - Independence
      - Normality
      - Equal variances

21. What are the different types of ANOVA tests
  - One-way ANOVA: one factor
  - Two-way ANOVA: two factors
  - Repeated measures ANOVA: repeated observations

22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test compares group variances.
  - It helps decide if group means differ significantly (e.g., in ANOVA).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm, ttest_ind, ttest_rel, chi2_contingency, chisquare
from statsmodels.stats.weightstats import ztest

# 1. Perform a Z-test comparing sample mean to population mean
data = np.random.normal(100, 10, 50)
z_score, p_value = ztest(data, value=105)
print("1. Z-test → z =", round(z_score, 2), ", p =", round(p_value, 4))

# 2. Simulate data for hypothesis testing and calculate p-value
sample = np.random.normal(98, 12, 40)
z = (np.mean(sample) - 100) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))
print("2. Simulated Z-test → z =", round(z, 2), ", p =", round(p, 4))

# 3. One-sample Z-test comparing mean with population mean
data_one_sample = np.random.normal(99, 8, 30)
z_stat, p_val = ztest(data_one_sample, value=100)
print("3. One-sample Z-test → z =", round(z_stat, 2), ", p =", round(p_val, 4))

# 4. Two-tailed Z-test visualization
critical_z = norm.ppf(0.975)
plt.axvline(-critical_z, color='red', linestyle='--', label='Critical -z')
plt.axvline(critical_z, color='red', linestyle='--', label='Critical +z')
plt.axvline(z_stat, color='blue', linestyle='-', label='Z-statistic')
plt.title("4. Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-score")
plt.legend()
plt.grid(True)
plt.show()

# 5. Visualize Type 1 and Type 2 Errors
def visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_alpha = norm.ppf(1 - alpha)
    critical_val = mu0 + z_alpha * se
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)
    plt.plot(x, y0, label="H0", color="blue")
    plt.plot(x, y1, label="H1", color="green")
    plt.axvline(x=critical_val, color='red', linestyle='--', label='Critical Value')
    plt.fill_between(x, 0, y0, where=(x > critical_val), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, y1, where=(x <= critical_val), color='orange', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("5. Type I and II Errors")
    plt.grid(True)
    plt.show()
visualize_errors()

# 6. Independent T-test
group1 = np.random.normal(75, 5, 30)
group2 = np.random.normal(78, 5, 30)
t_stat, p_val = ttest_ind(group1, group2)
print("6. Independent T-test → t =", round(t_stat, 2), ", p =", round(p_val, 4))

# 7. Paired T-test
before = np.random.normal(80, 4, 30)
after = before + np.random.normal(3, 1, 30)
t_stat, p_val = ttest_rel(before, after)
print("7. Paired T-test → t =", round(t_stat, 2), ", p =", round(p_val, 4))

# 8. Simulate Z-test and T-test, compare results
group_a = np.random.normal(100, 10, 25)
group_b = np.random.normal(103, 10, 25)
z_diff, p_z = ztest(group_a - group_b)
t_diff, p_t = ttest_ind(group_a, group_b)
print("8. Z-test: z =", round(z_diff, 2), ", p =", round(p_z, 4))
print("8. T-test: t =", round(t_diff, 2), ", p =", round(p_t, 4))

# 9. Confidence Interval for sample mean
def get_confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=se)
    return interval
ci = get_confidence_interval(group1)
print("9. 95% Confidence Interval:", ci)

# 10. Margin of Error
sample_std = np.std(group1, ddof=1)
z_critical = norm.ppf(0.975)
moe = z_critical * (sample_std / np.sqrt(len(group1)))
print("10. Margin of Error:", round(moe, 2))

# 11. Bayesian Inference using Bayes' Theorem
P_A = 0.01
P_B_given_A = 0.9
P_B_given_notA = 0.1
P_notA = 1 - P_A
P_B = P_B_given_A * P_A + P_B_given_notA * P_notA
P_A_given_B = (P_B_given_A * P_A) / P_B
print("11. Bayesian Inference → P(A|B) =", round(P_A_given_B, 4))

# 12. Chi-square test for independence
observed = np.array([[30, 10], [20, 40]])
chi2, p, dof, expected = chi2_contingency(observed)
print("12. Chi-square Test → χ² =", round(chi2, 2), ", p =", round(p, 4))

# 13. Expected frequencies for Chi-square test
print("13. Expected Frequencies:\n", expected)

# 14. Goodness-of-fit test
obs = np.array([16, 18, 16, 14, 12, 12])
exp = np.array([15, 15, 15, 15, 15, 15])
exp = exp * (obs.sum() / exp.sum())  # normalize
chi2_gof, p_gof = chisquare(f_obs=obs, f_exp=exp)
print("14. Goodness-of-Fit Test → χ² =", round(chi2_gof, 2), ", p =", round(p_gof, 4))
```
